## Cleaning data

```python
titanic_data.drop("Cabin", axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.dropna(inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
sns.heatmap(data=titanic_data.isnull(), yticklabels=False, cbar=False)
```

```python
sex=pd.get_dummies(titanic_data["Sex"], drop_first=True)
sex.head(5)
```

```python
embarked=pd.get_dummies(titanic_data["Embarked"],drop_first=True)
embarked.head(5)
```

```python
pcl=pd.get_dummies(titanic_data["Pclass"],drop_first=True)
pcl.head(5)
```

```python
titanic_data_new=pd.concat([titanic_data,sex,embarked,pcl],axis=1)
```

```python
titanic_data_new
```

```python
titanic_data_new.drop(["Sex","Name","Embarked","PassengerId","Ticket","Pclass"],axis=1,inplace=True)
```

```python
titanic_data_new.head(5)
```

## Train data

```python
#dependent variable and independent variable
X=titanic_data_new.drop(["Survived"],axis=1)
y=titanic_data_new["Survived"]
```

```python
#split train and test data
#from sklearn import cross_validation
from sklearn.model_selection import train_test_split
```

```python
#train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression(solver="lbfgs")
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
